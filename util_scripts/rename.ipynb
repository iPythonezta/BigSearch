This notebook will contain code to rename files present in Files\\raw dir

We will replace each hash in file's name with another

```python
import json
import os
```

```python
url_map_path = '../Data/url_map.json'
raw_files_path = '../Data/Files/raw/'
with open(url_map_path, 'r') as f:
    url_map = json.load(f)
```

```python
nonexistent_files = 0
ind = 0
ind_to_id_map = {}
for file_name in url_map.keys():
    file_path = os.path.join(raw_files_path, file_name+'.html')
    new_path = os.path.join(raw_files_path, str(ind)+'.html')
    if os.path.exists(file_path):
        os.rename(file_path, new_path)
        ind_to_id_map[ind] = file_name
        ind += 1
    else:
        nonexistent_files += 1
```

```python
print(f'Number of nonexistent files: {nonexistent_files}')
```

```python
with open('../Data/ind_to_id_map.json', 'w') as f:
    json.dump(ind_to_id_map, f)
```

```python
files = os.listdir(raw_files_path)
files_with_no_map = 0
for f in files:
    if len(f) > 10:
        os.remove(os.path.join(raw_files_path, f))
        files_with_no_map += 1
```
