```python
import os
import pandas as pd
```

```python
df_pg_rank_results = pd.read_csv('../Data/Page Rank Results/page_rank_output.csv', names=["Page URL ID", "Page Rank"])
```

```python
df_pg_rank_results.head()
```

```python
df_pg_rank_results.sort_values(by="Page Rank", ascending=False).reset_index(drop=True)
```

```python
num_pages = len(df_pg_rank_results) - 1
df_pg_rank_results['Score'] = 0
for i in range(10):
    start_idx = int(i * 0.1 * num_pages)
    end_idx = int((i + 1) * 0.1 * num_pages)
    df_pg_rank_results.loc[start_idx:end_idx, 'Score'] = 10 - i
```

```python
df_pg_rank_results.tail()
```

```python
import json

id_to_url = {}
with open('..\\Encodings\\Domain & URL encodings for page rank\\Page_rank_files\\id_to_url.json', 'r') as f:
    id_to_url = json.load(f)
```

```python
df_pg_rank_results["Page URL ID"] = df_pg_rank_results["Page URL ID"].astype(str)
df_pg_rank_results["URL"] = df_pg_rank_results["Page URL ID"].map(id_to_url)
df_pg_rank_results.head()
```

```python
df_pg_rank_results[["URL", "Page Rank", "Score"]].to_csv('..\\Page Rank Results\\page_rank_results_with_urls.csv', index=False)
```
